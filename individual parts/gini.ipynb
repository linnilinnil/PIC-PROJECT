```python
import pandas as pd
pi=pd.read_csv('pi.csv')
pi
```

```python
import numpy as np
pi.sort_values(by=['tot_doll'], ascending=True, inplace=True)
pi['percentile']=pi['tot_doll'].cumsum()/pi['tot_doll'].sum()
pi
```

```python
import matplotlib.pyplot as plt
plt.plot(np.linspace(0.,1.,len(pi)), pi['percentile'])
plt.plot([0,1], [0,1])
plt.show()
```

```python
import plotly.express as px
fig = px.line(pi, x=np.linspace(0.,1.,len(pi)), y="percentile",
              hover_name="tot_doll", width=800, height=600, 
              title="Lorenz Curve (illustrating inequality)")
fig.add_scatter(x=np.linspace(0.,1.,len(pi)), y=np.linspace(0.,1.,len(pi)),name="Equality Line")
fig.show()
```

From the Lorenz Curve, we can observe that the distribution of grants among candidates are not equal, as illustrated as not exactly lying on the Equality Line. For example, the top 20% total dollars is assigned to the percentile, while the top 80% total dollars is assigned to the percentile.
